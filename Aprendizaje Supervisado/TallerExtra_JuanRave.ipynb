0. Importar Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
```

1. Cargar datos

```python
nxl='/content/drive/MyDrive/Aprendizaje Supervisado/insurance.csv'
df=pd.read_csv(nxl)

# Convertir 'charges' a string(Texto), quitar el punto decimal y convertirlo a int(entero)
df['charges'] = df['charges'].astype(str).str.replace('.', '', regex=False).astype(int)

display(df.head())
```

```python
df.describe()
```

2. Ajustar las variables no numericas

```python
# Conversion de la categoria 'sex' a boleano(0=Hombre,1=Mujer)
df['sex'] = df['sex'].map({'male': 0, 'female': 1})

# Conversion de la categoria 'smoker' a boleano (0=no,1=si)
df['smoker'] = df['smoker'].map({'no': 0, 'yes': 1})

# Conversion de la variable 'region'
df = pd.get_dummies(df, columns=['region'], prefix='region', dtype=int)

display(df.head())
```

3. Particion de datos

```python
from sklearn.model_selection import train_test_split

X = df.drop('charges', axis=1)
y = df['charges']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

4. Modelo de regresion Sklearn

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Para usar las variables de la regresión, estas deben ser tipo array
X = np.array(X_train)
Y = np.array(y_train).reshape(-1, 1)

# Modelo de regresión
modelo = LinearRegression()
modelo.fit(X, Y)

# Ver coeficientes
print("Intercepto (a):", modelo.intercept_)
print("Pendiente (b):", modelo.coef_)
```

```python
df_train = pd.concat([y_train, X_train], axis=1)
corr_matrix = df_train.corr()
corr_matrix
```

5. Resultados y Metricas del modelo Train

```python
#Este es el metodo que vamos a utilizar siempre
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.datasets import make_regression
import statsmodels.api as sm

# Modelo base
lr = LinearRegression()

# Forward selection
sfs = SFS(lr,
          k_features='best',   # Puede ser número fijo o 'best'
          forward=True,        # True = forward, False = backward
          floating=False,      # True para stepwise mixto
          scoring='r2',
          cv=5)

sfs = sfs.fit(X_train, y_train)

print("Variables seleccionadas usando SequentialFeatureSelector (Forward):", sfs.k_feature_names_)

# Crear y ajustar el modelo OLS con las variables seleccionadas
X_train_selected = sm.add_constant(X_train[list(sfs.k_feature_names_)])
model_sfs = sm.OLS(y_train, X_train_selected).fit()

# Resumen del modelo OLS
print("\nResumen del modelo OLS con variables seleccionadas por SequentialFeatureSelector:")
print(model_sfs.summary())
```

5.1. Modelo con datos Test

```python
#Este es el metodo que vamos a utilizar siempre
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.datasets import make_regression
import statsmodels.api as sm

# Modelo base
lr = LinearRegression()

# Forward selection
sfs = SFS(lr,
          k_features='best',   # Puede ser número fijo o 'best'
          forward=True,        # True = forward, False = backward
          floating=False,      # True para stepwise mixto
          scoring='r2',
          cv=5)

sfs = sfs.fit(X_test, y_test)

print("Variables seleccionadas usando SequentialFeatureSelector (Forward):", sfs.k_feature_names_)

# Crear y ajustar el modelo OLS con las variables seleccionadas
X_test_selected = sm.add_constant(X_test[list(sfs.k_feature_names_)])
model_sfs = sm.OLS(y_test, X_test_selected).fit()

# Resumen del modelo OLS
print("\nResumen del modelo OLS con variables seleccionadas por SequentialFeatureSelector:")
print(model_sfs.summary())
```

6. Evaluacion de ambos modelos

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import statsmodels.api as sm

# Predicciones en el set de entrenamiento con el modelo entrenado con SFS features
X_train_selected = sm.add_constant(X_train[list(sfs.k_feature_names_)])
y_pred_train = model_sfs.predict(X_train_selected)

# RMSE
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
print(f"RMSE (Train): {rmse_train:.6f}")

# MAE
mae_train = mean_absolute_error(y_train, y_pred_train)
print(f"MAE (Train): {mae_train:.6f}")

# Calculate MAPE
# epsilon para evitar la division por cero
epsilon = 1e-10
mape_train = np.mean(np.abs((y_train - y_pred_train) / (y_train + epsilon))) * 100
print(f"MAPE (Train): {mape_train:.2f}%")

# R2
r2_train = r2_score(y_train, y_pred_train)
print(f"R2 (Train): {r2_train:.2f}")

# R2 ajustado
n_train = len(y_train) # numero de observaciones
p_train = X_train_selected.shape[1] - 1 # numero de predictores (excluyendo la constante)
adj_r2_train = 1 - (1 - r2_train) * (n_train - 1) / (n_train - p_train - 1)
print(f"R2 ajustado (Train): {adj_r2_train:.2f}")

# --- Calculando las métricas para el conjunto de prueba también (como antes) ---
print("\nMétricas en el conjunto de prueba:")
X_test_selected = sm.add_constant(X_test[list(sfs.k_feature_names_)])
y_pred_test = model_sfs.predict(X_test_selected)

# RMSE
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print(f"RMSE (Test): {rmse_test:.6f}")

# MAE
mae_test = mean_absolute_error(y_test, y_pred_test)
print(f"MAE (Test): {mae_test:.6f}")

# Calculate MAPE
mape_test = np.mean(np.abs((y_test - y_pred_test) / (y_test + epsilon))) * 100
print(f"MAPE (Test): {mape_test:.2f}%")

# R2
r2_test = r2_score(y_test, y_pred_test)
print(f"R2 (Test): {r2_test:.2f}")

# R2 ajustado
n_test = len(y_test) # numero de observaciones
p_test = X_test_selected.shape[1] - 1 # numero de predictores (excluyendo la constante)
adj_r2_test = 1 - (1 - r2_test) * (n_test - 1) / (n_test - p_test - 1)
print(f"R2 ajustado (Test): {adj_r2_test:.2f}")
```
